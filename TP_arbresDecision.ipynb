#Exercice 1 : 


```python
from sklearn import datasets
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# Charger l'ensemble de données à partir d'un fichier CSV
data = pd.read_csv('Tennis2.csv')

```

##Label encoding

```python
le = LabelEncoder()
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
```

```python
# Diviser les données en attributs et en sortie
X = data.drop('playTennis', axis=1)
#X = data.drop('Temperature1', axis=1)
y = data['playTennis']
```

##Création de l'arbre de décision

```python
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X, y)
```

##Représentation de l'arbre

```python
plt.figure(figsize=(20,10))
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['0', '1'])
plt.show()
```

##Matrice des confusions


```python
y_pred = clf.predict(X)
cm = confusion_matrix(y, y_pred)
print(cm)
```
